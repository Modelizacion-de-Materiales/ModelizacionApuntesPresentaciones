```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('dark_background')
plt.rc('figure',figsize=(10,10))
plt.rc('axes',labelsize=16)
plt.rc('xtick',labelsize=14)
plt.rc('ytick',labelsize=12)
plt.rc('legend',fontsize=12)
```

```python
X = np.linspace(-np.pi, np.pi,100)
Y = np.sin(X)
noise = (np.random.random(X.shape)-0.5)*1e-1
```

```python
plt.plot(X,Y,'--w',label='modelo')
plt.plot(X,Y+noise,'og',label='mediciones')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
```

En este ejemplo se representó una ley del tipo:
$$ Y = sin(X), -\pi < X < \pi $$
Se representa también un error en la medición con valores al azar con amplitud $10^{-1}$
